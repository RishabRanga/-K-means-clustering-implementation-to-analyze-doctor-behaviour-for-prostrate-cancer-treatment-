```python
import pandas as pd
```

```python
df=pd.read_csv("C:/Users/rishab/Desktop/835_837_NPI/835_837_NPI_Reduced.csv")
```

```python
feature=['procedure_code','age']
```

```python
df = df.fillna(0)
```

```python
import datetime
#procedure code null specifies no procedure was meant
now = datetime.datetime.now()
pat_age = []
pat_gender_code = []
pat_bday = df['a.pat_birthday']
_c22 = df['_c22']

for i in range(len(df)):
    pat_age.append(now.year - pat_bday.iloc[i])
blahdf = pd.DataFrame({'age' : pat_age})
blahdf.reset_index(drop=True, inplace=True)
df = df.join(blahdf)
```

```python
df
```

```python
#parsing the dx field for different types of cancer
dfC61=df[df['a.dx'].str.contains("C61")]
dfC67=df[df['a.dx'].str.contains("C67")]
dfE291=df[df['a.dx'].str.contains("E29.1")]
dfN200=df[df['a.dx'].str.contains("N20.0")]
dfN201=df[df['a.dx'].str.contains("N20.1")]
```

```python
dfC61
```

```python
x=dfC61[feature]
```

```python
import matplotlib
import numpy as np 
import matplotlib.pyplot as plt
```

```python
matplotlib.style.use('ggplot')
plt.figure(figsize=(14,7))
plt.subplot(1, 2, 1)
plt.scatter(x.age, x.procedure_code, s=40)
plt.xlabel("age")
plt.ylabel("procedure_code")
plt.title('Real Scatter before clusterinng')
plt.show()
```

```python
dfC61 = dfC61[["age", "procedure_code"]]
data = dfC61.as_matrix().astype("float32", copy = False)
from sklearn.preprocessing import StandardScaler
stscaler = StandardScaler().fit(data)
data = stscaler.transform(data)
from sklearn.cluster import DBSCAN
from sklearn import metrics
dbsc = DBSCAN(eps = .5, min_samples = 25).fit(data)
core_samples_mask = np.zeros_like(dbsc.labels_, dtype=bool)
#sets core sample indeces to true
core_samples_mask[dbsc.core_sample_indices_] = True

labels = dbsc.labels_

core_samples = np.zeros_like(labels, dtype = bool)

core_samples[dbsc.core_sample_indices_] = True
# Black removed and is used for noise instead.
unique_labels = set(labels)
colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = 'k'

    class_member_mask = (labels == k)

    xy = data[class_member_mask & core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k', markersize=14)

    xy = data[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k', markersize=6)

#plt.title('Estimated number of clusters: %d' % n_clusters_)
plt.show()
```
